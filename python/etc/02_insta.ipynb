```python
# case1 (mapì„ ì´ìš©)
def return_words(x):
    nouns = okt.nouns(x)
    if nouns:
        words = [n for n in nouns if len(n) > 1]
    else:
        words = ""
    return words
    

df['words'] = df['Commet'].map(return_words)
```

```python
df
```

```python
# case2 (forë¬¸ ì´ìš©)
col_data = []
for i in range(len(df)):
    # print(i)
    nouns = okt.nouns(df.loc[i, 'Commet'])
    if nouns:
        words = [n for n in nouns if len(n) > 1]
    else:
        words = ""
    # print(words)
    col_data.append(words)
df['words2'] = col_data
```

```python
# ì´ë¯¸ì§€ë¥¼ ì €ì¥í•˜ëŠ” í•¨ìˆ˜ë¥¼ í•˜ë‚˜ ìƒì„± 
def image_save(img_path, save_path, file_name):
    html_data = requests.get(img_path)
    imageFile = open(
        os.path.join(
            save_path, 
            file_name
        ), 
        'wb'
    )
    # ì´ë¯¸ì§€ ë°ì´í„°ì˜ í¬ê¸° 
    chunk_size = 100000000
    for chunk in html_data.iter_content(chunk_size):
        imageFile.write(chunk)
        imageFile.close()
    print('íŒŒì¼ ì €ì¥ ì™„ë£Œ')
```
